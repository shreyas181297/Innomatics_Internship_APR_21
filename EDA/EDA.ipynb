As our data does not include any missing values or redundant entries from initial inspection we can go ahead to perform Univariate analysis. Univariate analysis is perhaps the simplest form of statistical analysis. Like other forms of statistics, it can be inferential or descriptive

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,7))
sns.countplot(df['kills'].astype('str').sort_values())
plt.title("Kill Count",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=[10,6])
df['assists'].value_counts().plot(kind='bar',color="g")
plt.title("Distribution of assists") 
plt.ylabel("count") 
plt.show()
print(df['assists'].value_counts())
```

```python
df['boosts'].value_counts().plot(kind='bar')
plt.title("Distribution of boosts") 
plt.ylabel("count") 
plt.show()
print(df['boosts'].value_counts())
```

```python
plt.figure(figsize=[10,6])
sns.countplot(df['headshotKills'].sort_values())
plt.title("Distribution of headshotKills") 
plt.ylabel("count") 
plt.show()
```

```python
plt.figure(figsize=[10,6])
sns.countplot(df['DBNOs'].sort_values())
plt.title("Distribution of DBNOs") 
plt.ylabel("count") 
plt.show()
```

```python
plt.figure(figsize=[18,4])
sns.countplot(df['killPlace'].sort_values())
plt.title("Distribution of killPlace") 
plt.ylabel("count") 
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=[18,4])
sns.countplot(df['killStreaks'].sort_values())
plt.title("Distribution of killStreaks") 
plt.ylabel("count") 
plt.show()
```

```python
plt.figure(figsize=[18,4])
sns.countplot(df['maxPlace'].sort_values())
plt.title("Distribution of maxPlace") 
plt.ylabel("count") 
plt.show()
```

```python
plt.figure(figsize=[18,4])
sns.countplot(df['numGroups'].sort_values())
plt.title("Distribution of numGroups") 
plt.ylabel("count") 
plt.show()
```

As we can see from above two fraphs the number of groups in a match and max places is nearly similar because the total places will always depend upon the teams in a match. to confirm this relationship we will plot realtionship graph.

```python
plt.figure(figsize=(15,10))
ax1 = sns.boxplot(x="maxPlace",y="numGroups", data = df)
plt.title("maxPlace vs. Number of Groups")
plt.xticks(rotation=90)
plt.show()
```

Also from the previous observations and game understandings kills and damage dealt will be proportional to each other, Hence we will perform similar analysis to confirm this.

```python
plt.figure(figsize=(15,10))
ax1 = sns.boxplot(x="kills",y="damageDealt", data = df)
plt.title("Damage Dealt vs. Number of Kills")
plt.xticks(rotation=90)
plt.show()
```

As the figure mostly follows 45 degree slope we can confirm that the kills secured and damaged dealt are highly correalted

```python
plt.figure(figsize=(18,10))
plt.subplot(244)
sns.distplot(df['revives'])
plt.title("revives given")
plt.subplot(241)
sns.distplot(df['rankPoints'])
plt.title("Rank points obtained")
plt.subplot(242)
sns.distplot(df['winPoints'])
plt.title("Win points obtained")
plt.subplot(243)
sns.distplot(df['rideDistance'])
plt.title("Ride distance")
plt.show()
```
